```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
mpg = sns.load_dataset('mpg')
```

```python
#Looking at the relationship between Horsepower and Fuel Economy

plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.title('Relationship between Horsepower and Fuel Economy')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (mpg)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
#Looking at a table Summarizing Fuel Economy by Country of Origin

fuel_economy_by_origin = mpg.groupby('origin')['mpg'].mean().reset_index()
fuel_economy_by_origin.columns = ['Country of Origin', 'Average Fuel Economy (mpg)']
fuel_economy_by_origin
```

```python
import warnings
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Suppress the FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")

# Proceed with the rest of your code
plt.figure(figsize=(10, 6))
mpg['year'] = mpg['model_year'] + 1900
average_mpg_by_year = mpg.groupby('year')['mpg'].mean().reset_index()
sns.lineplot(x='year', y='mpg', data=average_mpg_by_year)
plt.title('Change in Fuel Economy Over Time')
plt.xlabel('Year')
plt.ylabel('Average Miles per Gallon (mpg)')
plt.grid(True)  # Add grid lines
plt.show()
```

```python
#looking at Distribution of Fuel economie 

plt.figure(figsize=(10, 6))
sns.histplot(mpg['mpg'], bins=20, kde=True)
plt.title('Distribution of Fuel Economy')
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='cylinders', y='mpg', data=mpg, palette="Set2")
plt.title('Fuel Economy by Number of Cylinders', fontsize=16)
plt.xlabel('Number of Cylinders', fontsize=14)
plt.ylabel('Miles per Gallon (mpg)', fontsize=14)

# Customize the grid lines
plt.grid(True, which='major', axis='y', linestyle='--', linewidth=0.7, color='gray')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

## Observations

1. **Inverse Relationship**: There is an inverse relationship between horsepower and fuel economy; as horsepower increases, fuel economy tends to decrease.
2. **Japanese Cars**: Japanese cars tend to have better fuel economy compared to American and European cars.
3. **Historical Improvement**: Fuel economy has generally improved over time, especially during the late 1970s and early 1980s.
4. **Fuel Economy Range**: Most cars have fuel economy between 15 and 30 mpg.
5. **Cylinders and Economy**: Cars with fewer cylinders tend to have better fuel economy.
